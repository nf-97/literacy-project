```python
import pandas as pd

df1 = pd.read_csv('/Users/liliamn/Desktop/merged_tables.csv')

df2 = pd.read_csv('/Users/liliamn/Desktop/gpi.csv')

# Examine the column names in each DataFrame
print("Column names in df1:", df1.columns)
print("Column names in df2:", df2.columns)


```

```python
df2['Country'] = df2['Country'].replace("North Korea", "Korea, Dem. People's Rep.")
df2
```

```python
merged_df = pd.merge(df1, df2, left_on='Country Name', right_on='Country', how='inner')

merged_df
```

```python
final_csv = merged_df.drop('Country', axis=1)

final_csv = final_csv.rename(columns={'Mean': 'Mean_GPI'})

final_csv
```
