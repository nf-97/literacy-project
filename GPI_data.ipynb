# Extracting Global Peace Index Data for our 10 countries of interest

The <b>Global Peace Index (GPI)</b> is an annual report published by the Institute for Economics and Peace (IEP) that measures the relative peacefulness of countries and regions worldwide. It evaluates indicators across three domains: societal safety and security, ongoing domestic and international conflicts, and level of militarization. Each country is assigned a score from 1 to 5, indicating its level of peace. The GPI helps policymakers and researchers understand global peace trends, identify areas for improvement, and promote conflict resolution. 

First, import the needed package:

```python
import pandas as pd
```

Then read the raw file into a new variable called df.

```python
df = "GPI_2022.xlsx"

df = pd.read_excel(df)
df
```

The Global Peace Index data is available from 2008 as GPI was launched in 2007. 

Define the list of countries.

```python
countries = ["Chad", "Afghanistan", "Mali", "Niger", "Guinea", "Cuba", "Lithuania", "Estonia", "Latvia", "North Korea"]
```

Filter the DataFrame only showing the data of our countries of interest and read it in a new variable called filtered_df.

```python
filtered_df = df[df["Country"].isin(countries)]

filtered_df
```

Print out the columns of the filtered DataFrame and drop all the unnecessary columns.

```python
filtered_df.columns
```

```python
filtered_df = filtered_df.drop(filtered_df.columns[[1, -1]], axis=1)

filtered_df
```

Reset the indicies. 

```python
filtered_df = filtered_df.reset_index(drop=True)
filtered_df
```

Calculate the mean of GPI for each country.

```python
filtered_df['Mean'] = filtered_df.iloc[:, 1:].mean(axis=1)
filtered_df
```

Create a new DataFrame from Countries and their corresponding mean GPI values.

```python
gpi = filtered_df[['Country', 'Mean']]
gpi
```

Visualise the final results and extract as a new file.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(gpi['Country'], gpi['Mean'], color='lightblue')
plt.xlabel('Country')
plt.ylabel('Mean')
plt.title('Mean GPI')
plt.xticks(rotation=90)
plt.show()
```

```python
csv_path = 'gpi.csv'

gpi.to_csv(csv_path, index=False)
```
