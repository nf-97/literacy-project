```python
import pandas as pd

file_path = "/Users/liliamn/PycharmProjects/cfg-python/GPI_2022.xlsx"

df = pd.read_excel(file_path)
df
```

The Global Peace Index data is available from 2008 as GPI was launched in 2007. 

```python
# Define the list of countries
countries = ["Chad", "Afghanistan", "Mali", "Niger", "Guinea", "Cuba", "Lithuania", "Estonia", "Latvia", "North Korea"]
```

```python
filtered_df = df[df["Country"].isin(countries)]

filtered_df
```

```python
filtered_df.columns
```

```python
filtered_df = filtered_df.drop(filtered_df.columns[[1, -1]], axis=1)

filtered_df
```

```python
filtered_df = filtered_df.reset_index(drop=True)
filtered_df
```

```python
filtered_df['Mean'] = filtered_df.iloc[:, 1:].mean(axis=1)
filtered_df
```

```python
gpi = filtered_df[['Country', 'Mean']]
gpi
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(gpi['Country'], gpi['Mean'], color='green')
plt.xlabel('Country')
plt.ylabel('Mean')
plt.title('Mean GPI')
plt.xticks(rotation=90)
plt.show()
```
